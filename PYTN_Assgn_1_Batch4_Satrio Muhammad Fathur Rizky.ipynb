### Merubah kolom borough menjadi country

```python
df.rename(columns={'borough':'country'}, inplace=True)
df.head()
```

### Melakukan Slicing terhadap Kolom yang dipakai

```python
df_country = df.loc[:,['country','major_category','minor_category','year','value']]
df_country.head()
```

### Melihat data kasus crime per country

```python
df_summed = df_country.groupby(['country'], as_index=False)['value'].sum()
df_summed
```

### Melihat kasus crime per tahun (major_catogory + minor_catogpry)

```python
df_sum_year = df.groupby(['year'], as_index=False)['value'].sum()
df_sum_year
```

### Menampilkan Line Plot 

```python
df_sum_year.set_index('year').plot(kind='line', figsize=(8,6), color='red', linewidth='2')
plt.title('LONDON CRIME 2008-2016')
plt.ylabel('Number of Crimes')
plt.xlabel('Years')
plt.show()
```

### Menampilkan top 5 country dengan jumlah crime tertinggi

```python
df_summed_sorted = df_summed.sort_values(by='value', ascending=False)
df_summed_sorted.head(5)
```

```python
df_top5 = df_summed_sorted.head(5)
df_top5.set_index('country', inplace = True)
df_top5
```

### Menampilkan Area Plot dari top 5 country dengan value crime tertinggi

```python
mpl.style.use(['ggplot'])
df_top5.plot(kind = 'area',
             stacked = False,
             figsize=(12,4),)
```

Dari Area plot ditunjukan bahwa kota WEstminster merupakan kota dengan jumlah kasus kriminal tertinggi selama tahun 2008 - 2016.

### Menampilkan area plot major_category pada country Westminster

```python
df_Westminster_areaPlot = Westminster[(Westminster['year'] > 2007)].groupby(['major_category','year'])['value'].sum()
df_Westminster_areaPlot.plot(kind='area', figsize=(30,5))
plt.xlabel('Year')
plt.ylabel('Number of crime')
plt.title('Major Category Area Westminster 2008-2016')
plt.show()
```

terlihat bahwa dari data major category untuk kota Westminster kasus yang sering terjadi adalah Theft and Handling.

### Menampilkan area plot minor_category pada country Westminster

```python
df_Westminster_areaPlot = Westminster[(Westminster['year'] > 2007)].groupby(['minor_category','year'])['value'].sum()
df_Westminster_areaPlot.plot(kind='area', figsize=(30,5))
plt.xlabel('Year')
plt.ylabel('Number of crime')
plt.title('Minor Category Area Westminster 2008-2016')
plt.show()
```

terlihat bahwa dari data minor category untuk kota Westminster kasus yang sering terjadi adalah other Theft Person.

### Menampilkan Major_Category = Robbery dari Tahun 2008 - 2016 

```python
major = df[(df['year'] > 2007)].groupby(['major_category','year'])['value'].sum()
robbery = major['Robbery']
robbery
df_robbery = robbery.to_frame()
df_robbery = df_robbery.rename(columns={'value': 'Robbery'})
df_robbery
```

### Menampilkan Major_Category = Burglary dari Tahun 2008 - 2016 

```python
burglary = major['Burglary']
burglary
df_burglary = burglary.to_frame()
df_burglary = df_burglary.rename(columns={'value': 'Burglary'})
df_burglary
```

### Menggabungkan 2 data frame

```python
df_burglary_robbery = pd.concat([df_robbery, df_burglary], axis=1)
df_burglary_robbery
```

### Menampilkan Area Plot perbandingan Burlgary dan Robbery dari Tahun 2008 - 2016

```python
df_burglary_robbery.plot(kind='area', alpha=0.35, figsize=(10,5))
plt.xlabel('Year')
plt.ylabel('Number of crime')
plt.title('Burglary and Robbery in London 2008 - 2016')
plt.show()
```

Terlihat data diatas bahwa hasil compare dari Burglary vs Robbery, selama tahun 2008 - 2016 Burglary masih jadi trend kasus crime di london.

```python
df_burglary_robbery.plot(kind='hist')
plt.title("histogram of Burglary and Robbery [2008-2016]")
plt.ylabel('number of years')
plt.xlabel('number of crimes')
plt.show()
```
